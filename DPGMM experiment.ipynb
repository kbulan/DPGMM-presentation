```python
import sklearn.mixture
import itertools
from sklearn import cross_validation
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn import datasets
from sklearn import mixture
from sklearn import linear_model
from sklearn import metrics
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
iris.data[0]
```

```python
X = iris.data
y = iris.target
```

```python
#fig = plt.figure()
color = ['red', 'greenyellow', 'blue']
for i in range(len(color)):
    xs = X[:, 0][y == i]
    ys = X[:, 1][y == i]
    plt.scatter(xs, ys, c=color[i])
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
#fig.add_subplot(231)
plt.show()
color = ['red', 'greenyellow', 'blue']
for i in range(len(color)):
    xs = X[:, 2][y == i]
    ys = X[:, 0][y == i]
    plt.scatter(xs, ys, c=color[i])
    plt.xlabel('Petal length')
    plt.ylabel('Sepal length')
#fig.add_subplot(232)
plt.show()
color = ['red', 'greenyellow', 'blue']
for i in range(len(color)):
    xs = X[:, 1][y == i]
    ys = X[:, 2][y == i]
    plt.scatter(xs, ys, c=color[i])
    plt.xlabel('Sepal width')
    plt.ylabel('Petal length')
plt.show()
color = ['red', 'greenyellow', 'blue']
for i in range(len(color)):
    xs = X[:, 0][y == i]
    ys = X[:, 3][y == i]
    plt.scatter(xs, ys, c=color[i])
    plt.xlabel('Sepal length')
    plt.ylabel('Petal width')
plt.show()
color = ['red', 'greenyellow', 'blue']
for i in range(len(color)):
    xs = X[:, 1][y == i]
    ys = X[:, 3][y == i]
    plt.scatter(xs, ys, c=color[i])
    plt.xlabel('Sepal width')
    plt.ylabel('Petal width')
plt.show()
color = ['red', 'greenyellow', 'blue']
for i in range(len(color)):
    xs = X[:, 2][y == i]
    ys = X[:, 3][y == i]
    plt.scatter(xs, ys, c=color[i])
    plt.xlabel('Petal length')
    plt.ylabel('Petal width')
plt.show()
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.3)
```

Gaussian Mixture Model

```python
gmm = mixture.GMM(n_components = 5)
```

```python
gmm.fit(X)
```

```python
predict = gmm.predict(X)
```

```python
predict
```

```python
y
```

```python
GMM_values = predict.copy()
```

```python
GMM_values[predict == 1] = 1
GMM_values[predict == 1] = 2
```

```python
GMM_values
```

```python
metrics.accuracy_score(GMM_values, y)
```

Dirichlet Process Gaussian Mixture Model

```python
dpgmm = mixture.DPGMM(n_components = 100, alpha = 100)
```

```python
dpgmm.fit(X)
```

```python
predict = dpgmm.predict(X)
```

```python
predict
```

```python
y
```
